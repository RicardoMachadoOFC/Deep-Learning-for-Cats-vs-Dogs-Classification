```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
print(os.listdir("data/train"))
```

```python
def preprocessar_imagem(caminho):
    img = Image.open(caminho)

    img = img.resize((64, 64))
    img = img.convert("L")

    img = np.array(img) / 255.0
    img = img.flatten()

    return img
```

```python
def carregar_dados():
    X = []
    y = []

    for img_name in os.listdir("data/train/cat"):
        path = "data/train/cat/" + img_name
        X.append(preprocessar_imagem(path))
        y.append(0)

    for img_name in os.listdir("data/train/dog"):
        path = "data/train/dog/" + img_name
        X.append(preprocessar_imagem(path))
        y.append(1)

    return np.array(X), np.array(y)


X, y = carregar_dados()
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)
```

```python
entrada = 4096
oculta = 128
saida = 1

np.random.seed(42)

W1 = np.random.randn(entrada, oculta) * 0.01
b1 = np.zeros((1, oculta))

W2 = np.random.randn(oculta, saida) * 0.01
b2 = np.zeros((1, saida))
```

```python
def forward(X):
    Z1 = X @ W1 + b1
    A1 = relu(Z1)

    Z2 = A1 @ W2 + b2
    A2 = sigmoid(Z2)

    return A1, A2
```

```python
def loss(y_true, y_pred):
    eps = 1e-8
    return -np.mean(
        y_true*np.log(y_pred + eps) +
        (1-y_true)*np.log(1-y_pred + eps)
    )
```

```python
A1, y_pred = forward(X)

print("Predições:", y_pred[:5])
print("Loss:", loss(y, y_pred))
```
